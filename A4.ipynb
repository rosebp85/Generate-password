```python
import customtkinter as ctk
import random
import string


def generate_random_password():
     
    password_length = int(entry_string.get()) 
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(password_length))
  
    insert_box.delete(0, ctk.END)  
    insert_box.insert(ctk.END, password) 

    val, score, errors = password_checker(password)
    a = password_strength(score)
    with open("password.txt", "a", encoding="utf-8") as file:
        file.write(f"{password}   - سطح امنیت: {a}\n")

    password_level(password)


def password_checker(passw):

    special_characters = ['$', '@', '#', '%', '!', '?', '^', '&', '*', ')', '(', '[', ']', '{', '}', '/', '+', '=', ':']
    val = True
    score = 0
    errors=[]
    

    if len(passw) < 6:
        errors.append('طول رمز باید حداقل 6 باشد')
        val = False
    else:
        score += 1



    if not any(char.isdigit() for char in passw):
        errors.append('باید حداقل دارای یک عدد باشد')
        val = False
    else:
        score += 1


    if not any(char.isupper() for char in passw):
        errors.append('باید حداقل دارای یک حرف بزرگ باشد')
        val = False
    else:
        score += 1


    if not any(char.islower() for char in passw):
        errors.append('باید حداقل دارای یک حرف کوچک باشد')
        val = False
    else:
        score += 1


    if not any(char in special_characters for char in passw):
        errors.append('باید حداقل دارای یک کاراکتر خاص باشد')
        val = False
    else:
        score += 1


    return val, score,errors
    


def password_strength(score):

    if score == 0:
        return "ضعیف"
    
    elif 1 <= score < 3:
        return "متوسط"
    
    elif 3 <= score < 5:
        return "خوب"
    
    elif score == 5:
        return "عالی" 


def password_level(password):

    #برای سنجش لول رمز تولید شده توسط برنامه 
    val, score,errors = password_checker(password)
    securityLevel = password_strength(score)
    text=(f"سطح امنیت: {securityLevel}")
    label.configure(text=text)

    if val==True:  
        securityLevel = password_strength(score)
        label.configure(text=f"سطح امنیت: {securityLevel}")  
    else:  
        errors_in_page = '\n'.join(errors) 
        label.configure(text=f"\n{errors_in_page}\n\n سطح امنیت: {securityLevel}")
        

def password_level_user():
    #برای سنجش رمزی که کاربر به صورت دستی وارد میکنه
    user_paswoord = user_entry_string.get()
    val, score, errors = password_checker(user_paswoord)

    if val==True: 
        securityLevel = password_strength(score)
        label2.configure(text=f"سطح امنیت: {securityLevel}") 
    else: 
        error_text = '\n'.join(errors)  
        securityLevel = password_strength(score)
        label2.configure(text=f"\n{error_text}\n\nسطح امنیت: {securityLevel}") 

    a = password_strength(score)
    with open("password.txt", "a", encoding="utf-8") as file:
        file.write(f"{user_paswoord}   - سطح امنیت: {a}\n")


root = ctk.CTk()
root.title("تولید رمز امن")

tabview = ctk.CTkTabview(root)
tabview.pack(padx=40, pady=40)

tabview.add("رمز تولید")
tabview.add("امنیت بررسی")
login_tab = tabview.tab("رمز تولید")
security_tab = tabview.tab("امنیت بررسی")


entry_label = ctk.CTkLabel(login_tab, text = 'کنید وارد زیر کادر در را نظر مورد رمز طول ')
entry_label.pack(pady=10)

entry_string = ctk.StringVar()  
entry = ctk.CTkEntry(login_tab, textvariable=entry_string)
entry.pack(pady=10)


button = ctk.CTkButton(login_tab, text="رمز تولید",command=generate_random_password)  
button.pack(pady=10)

insert_box = ctk.CTkEntry(login_tab, height=30, width=200)
insert_box.pack(pady=10)

label = ctk.CTkLabel(login_tab, text="امنیت سطح :")
label.pack(pady=20)



entry_label3 = ctk.CTkLabel(security_tab, text =  'بنویس رو خودت رمز')
entry_label3.pack(pady=10)

user_entry_string = ctk.StringVar()  
user_entry = ctk.CTkEntry(security_tab, textvariable= user_entry_string)
user_entry.pack(pady=10)

button2 = ctk.CTkButton(security_tab, text="امنیت بررسی", command=password_level_user)
button2.pack(pady=10)

label2 = ctk.CTkLabel(security_tab, text="امنیت سطح :")
label2.pack(pady=20)



root.geometry("700x600")
root.mainloop()    
```
